# Designing an Artificial Intelligence


It is already well known that the first step towards designing an artificial intelligence is generally obtaining and getting insights of the phenomena that is wanted to be controlled. Exploring previous records is important as it not only helps evaluating what data could be relevant or not, but also, it helps clear out if what is know applies to our case of study or design within a project. Having said that, it is also important that if a particular phenomenon is not entirely within our knowledge, to get experts in the topic as they will give a clear direction of where or what to explore first or create our first __hypotheses__. 

When disigning an __Artificial Intelligence__, as many other projects, there is a workflow that can readily help us maximising the likelyhood of achieving a complete autonomous functional system. In 99% of the times, an initial exploration of records can really tell us much of the necessary and relevant information we need in order to achieve a sufficiently accurate design. 

Those initial exloration steps are well known as the __Data Science Workflow__ and its main purpose is for you to get as many insights on the hardware, software and data that might be needed to further complete the complete autonomous agent. 







## The Data Science Workflow 


The __Data Science Worflow__ is defined of the series of steps needed in order to get an approximation of the first learning agent that can make sufficiently good predictions of a particular phenomenon. It is always accompained of initial __objectives or research questions definition__ and evaluation of potential data sources that might answer them. The figure below is a common Data science project workflow that is used as a guideline everytime a new project is to be developed. In this section, a brief description of the Data Science Workflow will be remembered. We hope this flow could be used everytime needed throughout the duration of any lab for reference.

### Objectives and Research Questions

As mentioned before, the first step towards the development of an AI is setting up the __objective or research question to investigate__. Clarity is key at this stage as any ambiguity could lead to confusion that will creep in at any point further away the process. You will like to try to avoid reframing everything once almost all work is done ðŸ˜“. 

Properly setting up the objective, is a fundamental step towards our design. However, more than only excerting the obvious this step is also useful to quantify and get into an agreement of what could be ambiguous terms. For example: if a particular agent needs information about the time of the day defined as morning, afternoon and night. How will we be segmenting the day per hour? Will morning be before 12:00 or as more colloquialy will finish before 11:00? Those are the kind of important definitions that will change the behaviour of the model if not set up in advance.  

### Data Capturing 

Once the objectives are clearly defined, the next step is to define where could we get an initial insight of the data that will help throughout our design. Getting an idea on which variables could be of use and where to find them is through mere __hypotheses__. If at the end, the data seem not to be conclusive, it is important to name alternative sources for the project not to be halted. 

### Manage and Clean Data

After obtaining the data we will work with, it is important to preprocess the data for it to become as reliable as prossible and in the best scale for the predictors. This step focuses in understanding the availability of the data, types of data, eliminating incomplete rows or completing them through imputation. The final goal is to have a cleaned or "complete" database to work with. Moreover, during this task, a separation from data observations is made for the purpose of validating our assumptions within the __Final Analysis__ step.

### Exploratory Data Analysis

Although __data engineering__ is defined as a complete field of study and depends on the task that needs to be investigated, few common practices can be defined as a basic approach for understanding the behaviour of the phenomenon. This are commonly named and form part of the __Exploratory Data Analysis__. Includes observing metrics such as the __units, range, average, dispersion, and type of data__ each of the measured values have. Once these are understood, we can follow to see possible distributions and interactions between the data. 

### Final Analysis

Once the phenomenon has been understood, a model is chosen that it is believed to cause its behaviour. For this process, a predictor or machine learning algorithm is used as a motor for our agent to learn from It. Furthermore, the chosen algorithm will be only trained and validated with one set of the data -more commonly known as __training set__. Once the model is trained and validated with an acceptable __performance metric__ we procede in testing the model to attempt predicting those observations within the __test set__; data the model has never seen.  More on this in later labs. 


### Reporting

Once the phenomenon is captured and validated, reporting the results becomes the final task of the project. Regardless of the obtained results, it is important to generate conlcusions about the __phenomenon, data and the capability of the model to capture its behaviour__. That is of great importance as in case results were conclusive, we will get to know more about the particular phenomenon. In case it was observed that there was no evidence towards understanding the dynamics of the research question, it is important to document possible reasons why that was the case. This will lead to other researchers or yourself to generate new hypotheses without having to revisit previous undocumented ones.

# Predicting Stator and Rotor Temperatures for an Electric Car Controller

## Introduction 

The main purpose of the data set's recording is to be able to model the stator and rotor temperatures of a PMSM in real-time. Due to the intricate structure of an electric traction drive, direct measurement with thermal sensors is not possible for rotor temperatures, and even in case of the stator temperatures, sensor outage or even just deterioration can't be administered properly without redundant modeling. In addition, precise thermal modeling gets more and more important with the rising relevance of functional safety.

Your task is to design a model with appropriate feature engineering, that estimates four target temperatures in a causal manner (no future values should be considered for actual predictions). In order to maintain real-time capability, model sizes should be as small as possible. Note that temperature estimation in production will be deployed on best-cost hardware of traction drives in an automotive environment, where lean computation and lightweight implementation is key. This exercise was taken from the __Kaggle__ comunity website. All the data and information regarding the research can be found here: https://www.kaggle.com/wkirgsn/electric-motor-temperature/tasks?taskId=352.

## Objectives

Investigate those variables that could more effectively predict the temperature within the components in the motor. That is: the __Stator yoke, winding, tooth and permanent magnet__ for failure prediction and maintainance. 

## Details

#### Target features:

- stator_yoke
- stator_winding
- stator_tooth
- pm

#### Input features:

The remaining features exclusive torque.
Torque is a quantity, which is not reliably measurable in field applications, such that this feature shall be omitted in this task.

First we can summarise the data of our experiments by physical figures such as __rms voltage and current, power, energy; among others__. That will give us a single number of used metric by experiment. Just as numpy arrays, we can vectorise operations in order to get these values in a single line!
